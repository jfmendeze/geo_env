```python
# Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pdb
import xarray as xr
```

```python
# Open the netCDF 
dset = xr.open_dataset('N21E039.SRTMGL1_NC.nc')
```

```python
# Selecting the variable the information about elevation
DEM = np.array(dset.variables['SRTMGL1_DEM'])
```

```python
# To visualize the data
plt.figure(figsize=(10,7))
plt.imshow(DEM)
cbar = plt.colorbar()
cbar.set_label('Elevation (m a.s.l) ',  fontsize = 15)
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)
plt.xticks(fontsize = 15) 
plt.yticks(fontsize = 15) 
plt.show
plt.savefig('assignment_1.png', dpi = 300)
```
